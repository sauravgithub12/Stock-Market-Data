```python
import pandas as pd
import pandas_datareader as pdr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 8,6
import seaborn as sb
sb.set()
%matplotlib inline
```

```python
df=pd.read_csv('NQ_4_hours_2020-12-31_2021-06-09_trades (1).csv')
```

```python
df.head()
```

```python
df=df.drop(['Unnamed: 0'], axis=1)
```

```python
df.head(50)
```

```python
df['ma_5'] = df['exit_price'].rolling(5).mean()
df['ma_20'] = df['exit_price'].rolling(20).mean()
df[19:25]
```

```python
df['signal'] = np.where(df['ma_5'] > df['ma_20'], 1, 0)
df['signal'] = np.where(df['ma_5'] < df['ma_20'], -1, df['signal'])
df.dropna(inplace=True)
df.head()
```

```python
df['return'] = np.log(df['exit_price']).diff()
df['system_return'] = df['signal'] * df['return']
df['entry'] = df.signal.diff()
df.head(50)
```

```python
df=df.dropna()
```

```python
df.head()
```

```python
#df.to_excel('final_file.xlsx')
```

```python
plt.rcParams['figure.figsize'] = 12, 6
plt.grid(True, alpha = .3)
plt.plot(df.iloc[0:]['exit_price'], label = 'df')
plt.plot(df.iloc[0:]['ma_5'], label = 'ma_5')
plt.plot(df.iloc[0:]['ma_20'], label = 'ma_20')
plt.plot(df[0:].loc[df.entry == 2].index, df[0:]['ma_5'][df.entry == 2], '^',
         color = 'g', markersize = 12)
plt.plot(df[0:].loc[df.entry == -2].index, df[:]['ma_20'][df.entry == -2], 'v',
         color = 'r', markersize = 12)
plt.legend(loc=2);
```

```python
plt.rcParams['figure.figsize'] = 12, 6
plt.plot(np.exp(df['return']).cumprod(), label='Buy/Hold')
plt.plot(np.exp(df['system_return']).cumprod(), label='System')
plt.legend(loc=2)
plt.grid(True, alpha=.01)
```
